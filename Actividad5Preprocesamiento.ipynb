5. ¿Tendrán los valores faltantes de `Weight` relación con alguna otra variable? Analízalo considerando las siguientes columnas:

Para ello deberás dividir cada columna en dos grupos: con/sin faltantes en `Weight` y utilizar métodos gráficos y la prueba t para ver si hay diferencias significativas. NO olvides incluir las conclusiones del análisis con los resultados obtenidos.

**NOTA**. Puedes reutilizar las funciones `Diagnose_MV_Numerical` y `Diagnose_MV_Categorical` definidas en Hands-On-Prep_Missing_Values.zip. Revisa nuevamente cómo utilizarlas.

```python
from scipy.stats import ttest_ind
def Diagnose_MV_Numerical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    labels=[]
    box_sr = pd.Series('',index = BM_MV.unique())
    for poss in BM_MV.unique():
        BM = BM_MV == poss
        box_sr[poss] = df[BM][str_att_name].dropna()
        labels.append(MV_labels[poss])
    figure= plt.figure(figsize=(6,3))
    plt.boxplot(box_sr,vert=False)
    plt.yticks([1,2],labels)
    plt.xlabel(str_att_name)
    plt.show()

    plt.figure(figsize=(8,3))

    att_range = (df[str_att_name].min(),df[str_att_name].max())

    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].hist()
        plt.xlim = att_range
        plt.xlabel(str_att_name)
        plt.title(MV_labels[poss])

    plt.show()

    group_1_data = df[BM_MV][str_att_name].dropna()
    group_2_data = df[~BM_MV][str_att_name].dropna()

    p_value = ttest_ind(group_1_data,group_2_data).pvalue

    print('p-value of t-test: {}'.format(p_value))
```

```python
from scipy.stats import chi2_contingency
def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    plt.figure(figsize=(10,4))
    for i,poss in enumerate(BM_MV.unique()):
        figure= plt.figure(figsize=(6,3))
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])
    p_value = chi2_contingency(contigency_table)[1]

    print('p-value of Chi_squared test: {}'.format(p_value))
```

```python
numerical_columns = ['Age', 'Number of siblings']
categorical_columns = ['Gender', 'Education']

BM_MV = response_df['Weight'].isna()
```

**Numerical Data**

```python
for columns in numerical_columns:
    print('Diagnosis of Missing Values for {}'.format(columns))
    Diagnose_MV_Numerical(response_df,columns,BM_MV)
    print('\n')
```

El p-value en las pruebas t realizadas son valores mayores a 5%, por lo tanto se acepta la hipótesis nula donde se afirma que las medias de los valores son iguales. Esto sugiere que los valores nulos de Weight no son causados o influenciados por "Age" ni "Number of Siblings".

**Categoroical Data**

```python
for columns in categorical_columns:
    print('Diagnosis of Missing Values for {}'.format(columns))
    Diagnose_MV_Categorical(response_df,columns,BM_MV)
    print('\n')
```

Los resultados de la prueba chi cuadrado para las variables Educación y Género mostraron valores p> 5%. Esto afirma que los valores nulos de la variable Weight no han sido influenciados por ninguna de las variables numéricas y categóricas analizadas anteriormente. Por lo tanto, los valores nulos de esta variable son MCAR.

6. Utiliza una medida de tendencia central para imputar los valores faltantes de `Weight`. ¿Cuál es recomendable y por qué?

```python
# Impute missing 'Weight' values using the median
response_df['Weight'].fillna(response_df['Weight'].median(), inplace=True)
```

Llenar los valores nulos con la mediana es lo más recomendado, ya que la variable peso está achatada hacia la izquierda. En este caso, la mediana es menos sensible a los outliers, por lo tanto la desición de escoger la mediana para llenar los valores nulos es más robusta.


7. ¿Hay valores atípicos de la variable `Weight`? Confírmalo con un gráfico de boxplot. ¿Cuántos son? Almacénalos en un dataframe independiente y elimínalos del conjunto original.

```python
plt.figure(figsize=(6,3))
response_df['Weight'].plot.box()
plt.title('Young People Weight')
plt.show()
```

```python
Q1 = response_df.Weight.quantile(0.25)
Q3 = response_df.Weight.quantile(0.75)
IQR = Q3-Q1
BM = (response_df.Weight > (Q3+1.5 *IQR)) | (response_df.Weight < (Q1-1.5 *IQR))
outliers= response_df[BM]
print(f"Hay {outliers.shape[0]} valores atípicos")
```

```python
response_df= response_df[~BM].reindex()
```
